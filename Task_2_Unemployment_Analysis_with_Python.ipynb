**Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading the dataset**

```python

data = pd.read_csv('/content/Unemployment in India.csv')
```

**Data cleaning**


```python
# Example: Converting a date column
data[' Date'] = pd.to_datetime(data[' Date'])
```

**Data visualization-Distribution of unemployment rate**

```python
plt.figure(figsize=(10,5))
sns.histplot(data[' Estimated Unemployment Rate (%)'], bins=30, kde=True)
plt.title('Distribution of Unemployment Rate')
plt.show()
```

**Data visualization-Correlation Heatmap**

```python
# Drop non-numeric columns before calculating correlations
numeric_data = data.select_dtypes(include=['number'])
plt.figure(figsize=(10,5))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
